### Q5. How can I create a pivot table that shows the mean `Area` and `Solidity` values for each bin class?

### Q6. Separate all beans into three types based on MajorAxisLength and create a new column named `MajorAxisLength_Type`

Type_1: `MajorAxisLength<250`

Type_2: `250<=MajorAxisLength<400`

Type_3: `MajorAxisLength>=400`



```python
# Separate all beans into three types based on MajorAxisLength and create a new column named `MajorAxisLength_Type`
# Type_1: `MajorAxisLength<250`
# Type_2: `250<=MajorAxisLength<400`
# Type_3: `MajorAxisLength>=400`

df['MajorAxisLength_Type'] = pd.cut(df['MajorAxisLength'], bins=[0, 250, 400, df['MajorAxisLength'].max()+1], labels=['Type_1', 'Type_2', 'Type_3'])
df.groupby('MajorAxisLength_Type').size()
```

### Q7. Drop all the rows labeled as `MajorAxisLength_Type=Type_1`

```python
# Drop all the rows labeled as `MajorAxisLength_Type=Type_1`
df2 = df[df['MajorAxisLength_Type'] != 'Type_1']
df2.groupby('MajorAxisLength_Type').size()
```

### Q9. Convert the `Class` using one-hot encoding and merge the new column into the original dataframe
